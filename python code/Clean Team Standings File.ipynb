# Cleaning NFL Team Standings for 2021 - 2022 Season

```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#read in csv file
standing = pd.read_csv("Standing.2021.csv")
standing.head()
```

```python
#create new dataframe with just season standings (exclude pre-season and post-season)
season = standing[standing["SeasonType"] == 1]
season.head()
```

```python
#create new column concatenating record for labeling purposes
columns = ['Wins', 'Losses', 'Ties']
season['Record'] = season[columns].apply(lambda row: ' - '.join(row.values.astype(str)), axis=1)
season.head()
```

```python
#export cleaned season dataframe
season.to_excel(r'C:\Users\leand\Downloads\Season Record.xlsx', sheet_name='Season', index = False)
```
